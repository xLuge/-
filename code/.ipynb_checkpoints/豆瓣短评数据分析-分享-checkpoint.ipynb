# 三十而已 豆瓣短评分析

## 数据读入

```python
# 导入所需包
import numpy as np
import pandas as pd
import re
import jieba

from pyecharts.charts import Pie, Bar, Map, Line, Page
from pyecharts import options as opts
```

```python
# 读入数据
df = pd.read_excel('../data/三十而已豆瓣短评7.22.xlsx')
print(df.shape) 
print(df.info()) 
```

```python
# 删除列
df = df.drop(['comment_time', 'page_url', 'votes_num'], axis=1)
df.head() 
```

```python
# 处理空值
df.fillna('无', inplace=True) 
```

```python
# 查看重复值和空值
print(df.duplicated().sum())
print(df.isnull().sum()) 
```

## 数据预处理

```python
# 字段扩充
def comment_label(x):
    if x == '力荐' or x == '推荐':
        return '好评'
    elif x == '还行':
        return '一般'
    else:
        return '差评'

df['comment_label'] = df['rating_num'].map(comment_label) 
df.comment_label.value_counts() 
```

```python
# 定义函数
def transform_rating(x):
    if x == '立荐':
        return '5星'
    elif x == '推荐':
        return '4星'
    elif x == '还行':
        return '3星'
    elif x == '较差':
        return '2星'
    elif x == '很差':
        return '1星'
    else:
        return '5星'  # 异常值使用众数替换

# apply函数
df['rating_num'] = df.rating_num.apply(transform_rating) 
df.rating_num.value_counts() 
```

## 数据可视化

### 500条热评评分分布

```python
rating_num = df.rating_num.value_counts() 
# 数据对
data_pair = [list(z) for z in zip(rating_num.index.tolist(), rating_num.values.tolist())]
data_pair
```

```python
# 绘制饼图
pie1 = Pie(init_opts=opts.InitOpts(width='1350px', height='750px'))
pie1.add('', data_pair=data_pair, radius=['35%', '60%'], rosetype='radius')
pie1.set_global_opts(title_opts=opts.TitleOpts(title='豆瓣500条热评评分分布'), 
                     legend_opts=opts.LegendOpts(orient='vertical', pos_top='15%', pos_left='2%'))
pie1.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{c} \n {d}%"))
pie1.set_colors(['#FF7F0E', '#1F77B4', '#2CA02C', '#D62728', '#946C8B'])
pie1.render() 
```

### 六位主演的提及次数

```python
df.head() 
```

```python
# 模式
pat_wangmanni = '漫妮|疏影'
pat_gujia = '顾佳|童瑶'
pat_zhongxiaoqin = '晓芹|晓彤'
pat_chenyu = '陈屿|杨玏'
pat_xuhuanshan = '幻山|泽锋'

# 计数
wangmanni_num = df.short_comment.str.contains(pat_wangmanni).sum()    
gujia_num = df.short_comment.str.contains(pat_gujia).sum()
zhongxiaoqin_num = df.short_comment.str.contains(pat_zhongxiaoqin).sum()
chenyu_num = df.short_comment.str.contains(pat_chenyu).sum()
xuhuanshan_num = df.short_comment.str.contains(pat_xuhuanshan).sum() 
```

```python
df_num = pd.DataFrame({
    'name': ['王漫妮', '顾佳', '钟晓芹', '陈屿', '许幻山'],
    'number': [wangmanni_num, gujia_num, zhongxiaoqin_num, chenyu_num, xuhuanshan_num]
})
df_num = df_num.sort_values('number', ascending=False)
df_num
```

```python
# 产生数据
x_data = df_num['name'].values.tolist()
y_data = df_num['number'].values.tolist()

# 条形图
bar1 = Bar(init_opts=opts.InitOpts(width='1350px', height='750px'))
bar1.add_xaxis(x_data)
bar1.add_yaxis('', y_data)
bar1.set_global_opts(title_opts=opts.TitleOpts(title='豆瓣热评中5位主演的提及次数'), 
                     visualmap_opts=opts.VisualMapOpts(max_=103))

bar1.render() 
```

### 豆瓣热评中六位主演的评价分布

```python
# 计算比例
def get_rate_num(pattern):
    rate_num = df[df.short_comment.str.contains(pattern)]['comment_label'].value_counts() 
    rate_num = round(rate_num / rate_num.sum()*100, 2) 
    try:
        haoping_rate.append(rate_num['好评']) 
    except:
        haoping_rate.append(0)
    try:
        chaping_rate.append(rate_num['差评'])
    except:
        chaping_rate.append(0)
    try:
        yiban_rate.append(rate_num['一般']) 
    except:
        yiban_rate.append(0)  

names = ['王漫妮', '顾佳', '钟晓芹', '陈屿', '许幻山']
haoping_rate = []
chaping_rate = []
yiban_rate = [] 

# 运行函数
get_rate_num(pattern=pat_wangmanni)
get_rate_num(pattern=pat_gujia)
get_rate_num(pattern=pat_zhongxiaoqin)
get_rate_num(pattern=pat_chenyu)
get_rate_num(pattern=pat_xuhuanshan) 
```

```python
df_rate = pd.DataFrame({
    'name': names,
    'haoping': haoping_rate,
    'chaping': chaping_rate,
    'yiban': yiban_rate
}) 
df_rate
```

```python
# 产生数据
x_data = df_rate['name'].values.tolist()
y_data1 = df_rate['haoping'].values.tolist()
y_data2 = df_rate['yiban'].values.tolist()
y_data3 = df_rate['chaping'].values.tolist()

# 条形图
bar2 = Bar(init_opts=opts.InitOpts(width='1350px', height='750px'))
bar2.add_xaxis(x_data) 
bar2.add_yaxis('好评', y_data1, stack='stack1')
bar2.add_yaxis('一般', y_data2, stack='stack1')
bar2.add_yaxis('差评', y_data3, stack='stack1')
bar2.set_global_opts(title_opts=opts.TitleOpts(title='豆瓣热评提及5位主演的评价分布'))
bar2.set_series_opts(label_opts=opts.LabelOpts(formatter='{c}%', position='inside'))
bar2.set_colors(['#1F77B4', '#2CA02C', '#D62728'])
bar2.reversal_axis()
bar2.render() 
```

```python
# 生成page
page = Page()
page.add(pie1, bar1, bar2) 
page.render('./html/豆瓣短评数据分析.html') 
```

### 豆瓣好评词云图

```python
# 定义分词函数
def get_cut_words(content_series):
    # 读入停用词表
    stop_words = [] 
    
    with open(r"C:\Users\wzd\Desktop\CDA\CDA_Python\Python文本分析\10.文本摘要\stop_words.txt", 'r', encoding='utf-8') as f:
        lines = f.readlines()
        for line in lines:
            stop_words.append(line.strip())

    # 添加关键词
    my_words = ['王漫妮', '顾佳', '钟晓芹', '陈屿', '许幻山',
                '江疏影', '童瑶', '毛晓彤', '杨玏', '李泽锋']
    
    for i in my_words:
        jieba.add_word(i) 

    # 自定义停用词
    my_stop_words = ['真的', '这部', '这是', '一种', '一段']  
    stop_words.extend(my_stop_words)               

    # 分词
    word_num = jieba.lcut(content_series.str.cat(sep='。'), cut_all=False)

    # 条件筛选
    word_num_selected = [i for i in word_num if i not in stop_words and len(i)>=2]
    
    return word_num_selected
```
